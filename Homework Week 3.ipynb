```python
import pandas as pd
from random import randint
```

```python
def daymiles(trip_distance):
    total_distance = 0
    days = 1
    log = {}
    log["Day 0"] = 0
    
    while trip_distance > 0:
        day_distance = randint(50, 70)
        if day_distance >= trip_distance:
            day_distance = trip_distance
        log[f'Day {days}'] = day_distance + total_distance
        days += 1
        total_distance += day_distance
        trip_distance -= day_distance
    
    return log
        
```

```python
odometer = daymiles(1518)
# odometer_pd = pd.DataFrame(data = odometer, columns=['Day', 'Miles'])
odometer_pd = pd.Series(odometer)
log_pd = odometer_pd.diff()
#print(odometer_pd)
print(log_pd)
```
